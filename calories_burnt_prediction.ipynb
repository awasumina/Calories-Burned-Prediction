```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
calories = pd.read_csv('/content/calories.csv')
```

```python
exercise_data = pd.read_csv('/content/exercise.csv')
```

```python
calories_data = pd.concat([exercise_data, calories['Calories']], axis=1)
```

```python
# Converting the text data to numerical values
calories_data.replace({"Gender":{'male':0,'female':1}}, inplace=True)
```

```python
calories_data.head()
```

```python
X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# XGBoost Regressor
model = XGBRegressor()
```

```python
feature_importances = model.feature_importances_
sns.barplot(x=feature_importances, y=X.columns)
plt.title("Feature Importance")
plt.show()
```

```python
model.fit(X_train, Y_train)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
# compare actual and predicted values
comparison_df = pd.DataFrame({
    'Actual Calories': Y_test.values,
    'Predicted Calories': test_data_prediction,
    'Error (Difference)': Y_test.values - test_data_prediction
})

print(comparison_df.head(15))

plt.figure(figsize=(10, 6))
plt.scatter(Y_test, test_data_prediction, color='blue', alpha=0.5)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--')  # Line for perfect prediction
plt.title('Actual vs Predicted Calories')
plt.xlabel('Actual Calories')
plt.ylabel('Predicted Calories')
plt.grid()
plt.xlim(0, Y_test.max() + 50)
plt.ylim(0, Y_test.max() + 50)
plt.show()

```

```python
residuals = Y_test - test_data_prediction
plt.figure(figsize=(10,6))
sns.scatterplot(x=test_data_prediction, y=residuals)
plt.axhline(0, linestyle='--', color='red')
plt.xlabel("Predicted Calories")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(Y_test.values[:50], label='Actual Calories', marker='o', color='blue')
plt.plot(test_data_prediction[:50], label='Predicted Calories', marker='x', color='orange')
plt.title("Actual vs Predicted Calories (First 50 values)")
plt.xlabel("Index")
plt.ylabel("Calories")
plt.legend()
plt.show()
```

```python
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
print("Mean Absolute Error = ", mae)
```
